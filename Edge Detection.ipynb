# Edge detection using Sobel and zero crossing detector

### Algorithm:
1. Decalare the filters
2. Use convolution on the image to apply filters
3. Display the image

```python
# Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from scipy.ndimage import median_filter
import matplotlib.gridspec as gridspec
```

```python
image = cv2.imread('lena.jpg',0)
plt.imshow(image,cmap='gray')
```

```python
sobel_x = np.array([[-1,0,1],
                   [-2,0,1],
                   [-1,0,1]])

sobel_y = np.array([[-1,-2,-1],
                   [0,0,0],
                   [1,2,1]])

laplacian = np.array([[1,1,1],
                     [1,-8,1],
                     [1,1,1]])
```

```python
# apply the convolution operation bw image and each filter
sobel_x_filtered_img = convolve2d(image, sobel_x, mode='same', boundary='symm')
sobel_y_filtered_img = convolve2d(image, sobel_y, mode='same', boundary='symm')
laplacian_filtered_img = convolve2d(image, laplacian, mode='same', boundary='symm')
```

```python
# Plot the figures
plt.figure(figsize=(20,20))
gs = gridspec.GridSpec(4, 2)
gs.update(wspace=0.025, hspace=0.1)

fig = plt.subplot(gs[0])
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

fig = plt.subplot(gs[1])
plt.imshow(sobel_x_filtered_img, cmap='gray')
plt.title("Sobel Filter X")
plt.axis('off')

fig = plt.subplot(gs[2])
plt.imshow(sobel_y_filtered_img, cmap='gray')
plt.title("Sobel Filter Y")
plt.axis('off')

fig = plt.subplot(gs[3])
plt.imshow(laplacian_filtered_img, cmap='gray')
plt.title("Laplacian Filter")
plt.axis('off')

plt.show()
```
