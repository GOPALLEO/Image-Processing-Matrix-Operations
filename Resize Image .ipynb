# Using simple matrix operations to reduce the size of image
# 220X220 image reduced to 50% 90% and 80%
### Algorithm 
 1. For 50% reduction, skip every 2nd row and column to generate a new image
 2. For 90% reduction, take every 10th row and coulmn to generate new image (skipping the 9 in between)

```python
# Importing the libraries
# cv2 for loading the image matrix
# numpy for numerical calculations
# matplotlib to plot the results
import cv2
import numpy as np
from matplotlib import pyplot as plt 
```

```python
# Reading and displaying the image dimensions (Grayscale)
image = cv2.imread('lena.jpg',0)
print(image.shape)
```

```python
# To convert the image to other color can uncomment it (By default opencv reads image in BGR format)
# cv2.COLOR_BGR2GRAY
# cv2.COLOR_BGR2HSV
# cv2.COLOR_BGR2RGB

# image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY )
plt.imshow(image, cmap='gray')
```

```python
# Reduce every alternate row and column by half from the image resulting in decrease in dimension
reduced_50_percent = image[::2,::2]
reduced_50_percent.shape
```

```python
# Display the resultant reduced image to 50% of the orignal size
plt.imshow(reduced_50_percent, cmap='gray')
```

```python
# Take every 10th pixel into consideration - reducing the image to 
reduced_90_percent = image[::10,::10]
reduced_90_percent.shape
```

```python
plt.imshow(reduced_90_percent, cmap='gray')
```

```python
# remove every 5th row and column for reduction in the image to 80% of its orignal size and display the resulting shape
temp_image         = np.delete(image, list(range(0, image.shape[0], 5)), axis=0)
reduced_80_percent = np.delete(temp_image, list(range(0, image.shape[1], 5)), axis=1)
reduced_80_percent.shape
```

```python
plt.imshow(reduced_80_percent, cmap='gray')
```

```python
# Bonus - Calculation of percentage size reduced by the image
before=220
after=22
resize_percentage = ((after - before)/before)*100
print(f"The percentage of reduction is {resize_percentage}")
```
