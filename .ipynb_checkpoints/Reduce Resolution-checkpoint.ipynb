# Reduce and Display Image intensity from 8 bit to 1 in steps
### Algorithm
    1. Divide each pixel value if the 'n' bit orignal image by 2^n
    2. For converting to a 'k' bit image, multiply the result by 2^k-1 and round of to integer
    3. Make the negative values 0 after rounding off(Approximate)
    4. Display the images
    
Image containing 256 values - values that vary from 1 to 256 
but as we take it from 0 - values are from 0 to 255

An 8 bit number is 0000 0000
where each place denotes the power of 2

    binary value 0000 0000
    2^n      n = 7654 3210
    
1 0 1
2 00 2
3 000 8
4 0000 16
5 00000 32
6 000000 64
7 0000000 128

Total - 256 values
Removing each bit will remove the ability to contain higher values of numbers 


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_8bit = cv2.imread('lena.jpg',0)
print(np.max(image_8bit))
np.unique(image_8bit)
```

```python
plt.imshow(image_8bit,cmap='gray')
```

```python
# convert to 7 bit image using the algorithm n=8 k=7
image_7bit = np.round((image_8bit/256)*128-1)
image_7bit[image_7bit < 0] =0
np.unique(image_7bit)
```

```python
plt.imshow(image_7bit, cmap='gray')
```
